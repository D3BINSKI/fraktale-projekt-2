# Ruch Browna (proces Wienera)



# Drogi ruchu Browna

## Własności

**1. Ruch Losowy**

Droga ruchu Browna jest przykładem ruchu losowego, gdzie położenie cząsteczki w każdym kolejnym momencie czasu jest nieprzewidywalne i zależy od przypadkowych fluktuacji.

**2. Zależność od Czasu**

W ruchu Browna, przemieszczenie cząsteczki jest funkcją czasu, ale ta zależność nie jest deterministyczna.

**3. Niezależne Przyrosty**

W procesie Wienera, który jest matematycznym modelem ruchu Browna, przyrosty są statystycznie niezależne. Oznacza to, że przemieszczenie cząsteczki w jednym przedziale czasowym jest niezależne od jej przemieszczenia w innym przedziale.

**4. Normalny Rozkład Przyrostów**

Przyrosty w ruchu Browna mają rozkład normalny (Gaussa) z średnią zero i wariancją proporcjonalną do długości przedziału czasowego.

**5. Ciągłość**

Ścieżka ruchu Browna jest ciągła w czasie, co oznacza, że nie ma nagłych skoków w trajektorii cząsteczki.

**6. Brak Skali Charakterystycznej**

W ruchu Browna nie istnieje charakterystyczna skala długości ani czasu. Zachowanie ruchu Browna jest samo-podobne, co oznacza, że wygląda podobnie na wszystkich skalach czasowych.

**7. Ruch Niezrównoważony**

Ruch Browna nie dąży do osiągnięcia stanu równowagi.

**8. Dyfuzja**

W skali makroskopowej, ruch Browna przyczynia się do zjawiska dyfuzji, gdzie cząsteczki rozprzestrzeniają się z obszarów o wysokim stężeniu do obszarów o niższym stężeniu.

```python
import numpy as np
import matplotlib.pyplot as plt

def wiener_process(n, delta_t=1):
    """Generuje trajektorię procesu Wienera z n punktów."""
    # Generowanie n przyrostów z rozkładu normalnego
    dw = np.sqrt(delta_t) * np.random.randn(n-1)
    # Dodawanie początkowego punktu (0) i obliczanie trajektorii
    W = np.concatenate([[0.], np.cumsum(dw)])
    return W

# Generowanie trajektorii procesu Wienera
# wiener_process = np.cumsum(np.sqrt(delta_t) * np.random.randn(n))
# T = np.linspace(0, 1, n)

def plot_2D(W, tytul):
    # Rysowanie trajektorii procesu Wienera
    n = len(W)
    plt.figure(figsize=(10, 6))
    plt.plot(np.linspace(0, 1, n), W)
    plt.title(tytul)
    plt.xlabel("t")
    plt.ylabel("W(t)")
    plt.grid(True)
    plt.show()

n = 5000
delta_t = 1/n
W_1 = wiener_process(n, delta_t)
W_2 = wiener_process(n, delta_t)
```

```python
plot_2D(W_1, "Trajektoria procesu Wienera")
plot_2D(W_2, "Trajektoria procesu Wienera")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def plot_3D(t, X_t, Y_t, tytul_x, tytul_y):

    # Tworzenie wykresu 3D
    fig = plt.figure(figsize=(14, 8))
    
    ax = fig.add_subplot(111, projection='3d')
    
    # Rysowanie wykresu
    ax.plot(t, X_t, Y_t)

    # Etykiety osi
    ax.set_xlabel('t')
    ax.set_ylabel(tytul_x)
    ax.set_zlabel(tytul_y)

    # Wyświetlenie wykresu
    plt.show()
    
    return
```

```python
n = 1000
delta_t = 1/n
W_1 = wiener_process(n, delta_t)
W_2 = wiener_process(n, delta_t)

t = np.linspace(0, 1, n)
plot_3D(t, W_1, W_2, tytul_x='proces Wienera X(t)', tytul_y='proces Wienera Y(t)')
```

# Ułamkowe ruchy Browna (fBm)

# Droga fBm


Droga fBm (ułamkowego ruchu Browna) jest uogólnieniem klasycznego ruchu Browna (procesu Wienera) i została wprowadzona przez Benoita Mandelbrota oraz Johna Van Ness w 1968 roku. Ułamkowy ruch Browna charakteryzuje się kilkoma kluczowymi właściwościami, które odróżniają go od standardowego ruchu Browna.

## Własności drogi fBm

1. **Samopodobieństwo (Self-similarity)**

fBm jest samopodobny w skali Hursta $H$, co oznacza, że zmieniając skalę czasu o czynnik $\lambda$, trajektoria zmienia się jak $\lambda^H$ w skali przestrzennej. Dla $H = 1/2$ fBm jest ruchem Browna.

2. **Zależność Długoterminowa (Long-range Dependence)**

Dla $H > 1/2$, istnieje dodatnia korelacja między przyrostami, co oznacza, że kierunek przeszłego ruchu ma wpływ na przyszły kierunek ruchu. Dla $H < 1/2$, korelacja jest ujemna, co oznacza, że przeszły ruch przeciwny sprzyja przyszłemu ruchowi przeciwnym.

3. **Niezależność Przyrostów**

fBm nie posiada niezależnych przyrostów, co jest zasadniczą różnicą w porównaniu do klasycznego ruchu Browna. W fBm, przyrosty są zależne. Wyjątkiem jest ruch Browna dla $H = 1/2$.

4. **Rozkład Przyrostów**

Przyrosty fBm mają rozkład Gaussa, podobnie jak w przypadku klasycznego ruchu Browna, ale ich wariancja i średnia różnią się.

5. **Niezrównoważony Ruch**

fBm, podobnie jak ruch Browna, jest procesem niestacjonarnym i nie osiąga stanu równowagi.

6. **Kontinuum**

Droga fBm jest ciągła, ale jej pochodna nie musi być ciągła.

7. **Parametr Hursta**

Kluczowym parametrem w fBm jest parametr Hursta $H$, który leży w przedziale $(0,1)$ i określa charakterystykę samopodobieństwa oraz zależności długoterminowej procesu.

8. **Zastosowania**

fBm znajduje zastosowanie w wielu dziedzinach, w tym w finansach (do modelowania niektórych rodzajów rynków finansowych), w hydrologii (do modelowania przepływów rzek), w fizyce i np. do modelowania ruchu internetowego. 


Powyższe własności sprawiają, że fBm jest narzędziem często wykorzystywanym do modelowania procesów z zależnościami długoterminowymi i samopodobieństwem, które są niewystarczająco opisane przez standardowy ruch Browna.

# Proces Levy'ego



Proces Lévy'ego jest procesem stochastycznym, który jest naturalnym uogólnieniem procesu Wienera (ruchu Browna) i procesu Poissona. Proces Lévy'ego jest procesem stochastycznym z "ciężkimi ogonami", co oznacza, że ma on wysokie prawdopodobieństwo dużych skoków.

https://en.wikipedia.org/wiki/L%C3%A9vy_process

Charakterystyczne dla procesów Lévy'ego jest to, że mogą one zawierać zarówno skoki, jak i ciągłe ścieżki. Klasycznym przykładem procesu Lévy'ego jest proces Wienera i proces Poissona, ale istnieje wiele innych, bardziej skomplikowanych procesów Lévy'ego. 

Procesy Lévy'ego znajdują zastosowanie w wielu dziedzinach, w tym w matematyce finansowej, teorii kolejek, fizyce i biologii, gdzie modelują różnego rodzaju zjawiska o charakterze losowym z możliwością gwałtownych zmian (skoków).

```python
import numpy as np

def generate_levy_trajectory(n, alpha, beta):
    """
    Generuje trajektorię procesu Lévy'ego.
    
    :param n: Liczba kroków.
    :param alpha: Parametr skali (stabilności), 0 < alpha <= 2.
    :param beta: Parametr asymetrii, -1 <= beta <= 1.
    :return: Trajektoria procesu Lévy'ego.
    """
    if not (0 < alpha <= 2):
        raise ValueError("Parametr alpha musi być w zakresie (0, 2].")

    if not (-1 <= beta <= 1):
        raise ValueError("Parametr beta musi być w zakresie [-1, 1].")

    # Generowanie skoków z rozkładu Lévy'ego
    sigma = (1 - beta ** 2) ** (1 / (2 * alpha))
    uniform_randoms = np.random.uniform(-np.pi / 2, np.pi / 2, n)
    exponential_randoms = np.random.exponential(scale=(1 / alpha), size=n)
    jumps = sigma * np.tan(alpha * uniform_randoms) + beta * sigma * \
            (np.cos(alpha * uniform_randoms) - 1) / np.cos(uniform_randoms) / exponential_randoms

    # Sumowanie skoków do utworzenia trajektorii
    trajectory = np.cumsum(jumps)

    return trajectory
```

```python
# wybór parametrów

n = 1000
alpha = 1.5
beta = 0
t = np.linspace(0, n, n)
```

```python
levy_trajectory_1 = generate_levy_trajectory(n, alpha, beta)
plot_2D(levy_trajectory_1, 'Trajektoria procesu Lévy\'ego')

levy_trajectory_2 = generate_levy_trajectory(n, alpha, beta)
plot_2D(levy_trajectory_2, 'Trajektoria procesu Lévy\'ego')

plot_3D(t, levy_trajectory_1, levy_trajectory_2, 'Proces Lévy\'ego', 'Proces Lévy\'ego')
```
